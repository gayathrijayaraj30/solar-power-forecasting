```python
# 01_EDA.ipynb

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set visualization style
sns.set(style="whitegrid")

# Load datasets (update paths if needed)
inverter_data_path = "../data/inverter.csv"
plant_data_path = "../data/plant.csv"

inverter_df = pd.read_csv(inverter_data_path, parse_dates=["Updated Time"])
plant_df = pd.read_csv(plant_data_path, parse_dates=["Updated Time"])

# Display first few rows to check
print("Inverter Data Sample:")
display(inverter_df.head())

print("Plant Data Sample:")
display(plant_df.head())
```

```python
# Summary statistics
print("\nInverter Data Summary:")
display(inverter_df.describe())

print("\nPlant Data Summary:")
display(plant_df.describe())
```

```python
# Check for missing values
print("\nMissing values in Inverter Data:")
print(inverter_df.isnull().sum())

print("\nMissing values in Plant Data:")
print(plant_df.isnull().sum())
```

```python
# Plot time series for Production, Consumption for inverter data
plt.figure(figsize=(14,6))
plt.plot(inverter_df["Updated Time"], inverter_df["Production(kWh)"], label="Production (kWh)")
plt.plot(inverter_df["Updated Time"], inverter_df["Consumption(kWh)"], label="Consumption (kWh)")
plt.title("Inverter: Production and Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("kWh")
plt.legend()
plt.show()
```

```python
# Plot time series for Plant data: Production-Today, Consumption-Today
plt.figure(figsize=(14,6))
plt.plot(plant_df["Updated Time"], plant_df["Production-Today(kWh)"], label="Production-Today (kWh)")
plt.plot(plant_df["Updated Time"], plant_df["Consumption-Today(kWh)"], label="Consumption-Today (kWh)")
plt.title("Plant: Production and Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("kWh")
plt.legend()
plt.show()
```

```python
# Correlation heatmap for inverter data numeric columns
plt.figure(figsize=(8,6))
sns.heatmap(inverter_df.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm")
plt.title("Inverter Data Correlation Matrix")
plt.show()
```

```python
# Correlation heatmap for plant data numeric columns
plt.figure(figsize=(8,6))
sns.heatmap(plant_df.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm")
plt.title("Plant Data Correlation Matrix")
plt.show()
```

```python
# Compare Production from both datasets on same plot
plt.figure(figsize=(14,6))
plt.plot(inverter_df["Updated Time"], inverter_df["Production(kWh)"], label="Inverter Production")
plt.plot(plant_df["Updated Time"], plant_df["Production-Today(kWh)"], label="Plant Production")
plt.title("Comparison: Inverter vs Plant Production")
plt.xlabel("Date")
plt.ylabel("kWh")
plt.legend()
plt.show()
```
